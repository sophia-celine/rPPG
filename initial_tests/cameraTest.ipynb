```python
import numpy as np
import cv2
import glob
import os

width = 2464
height = 2048
fps = 25
folder = r"C:\Users\Sophia\Pictures\Baumer Image Records\VCXG.2-57C\fpstest"

frame_files = sorted(glob.glob(os.path.join(folder, "*.brw")))

frames = []
for file in frame_files:
    with open(file, "rb") as f:
        raw = np.frombuffer(f.read(), dtype=np.uint8)[235:]
    
    bayer = raw.reshape((height, width))

    rgb = cv2.cvtColor(bayer, cv2.COLOR_BayerRG2BGR)

    frames.append(rgb)
```

```python
out = cv2.VideoWriter(
    "output.avi",
    cv2.VideoWriter_fourcc(*"XVID"),
    fps,
    (width, height)
)

for frame in frames:
    out.write(frame)

out.release()
```
